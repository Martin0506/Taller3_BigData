## Introducción

En esta sección, exploraremos cada uno de los mini-datasets de prueba para determinar el tipo de visualización más adecuado para los datos, según el problema específico a abordar.  

Para el dataset de campañas de marketing, nos interesa examinar las posibles relaciones entre los ingresos por ventas y dos variables clave:
- El dinero invertido en la campaña  
- La duración de la campaña  

Al seleccionar el tipo de gráfico apropiado para cada relación, podremos comprender mejor los patrones subyacentes y obtener información valiosa a partir de los datos.



```python
import pandas as pd

# Read the marketing campaigns dataset
marketing_df = pd.read_csv("marketing_campaigns.csv")

# Display the first few rows for exploration
print(marketing_df.head())
print(marketing_df.info())
print(marketing_df.describe())
```

```python
import matplotlib.pyplot as plt
import numpy as np

# === Scatter + Trendline: Sales Revenue vs. Ad Spend ===
x = marketing_df['Ad_Spend_USD']
y = marketing_df['Sales_Revenue_USD']

plt.figure(figsize=(6,4))
plt.scatter(x, y, label="Datos")

# Calcular línea de tendencia (regresión lineal)
m, b = np.polyfit(x, y, 1)   # pendiente y intercepto
plt.plot(x, m*x + b, color="red", label="Tendencia")

plt.xlabel("Ad Spend (USD)")
plt.ylabel("Sales Revenue (USD)")
plt.title("Sales Revenue vs. Ad Spend")
plt.legend()
plt.grid(True)
plt.show()


```

El gráfico de dispersión muestra la relación entre el gasto en publicidad (**Ad Spend**) y los ingresos obtenidos en ventas (**Sales Revenue**).  

Cada punto representa una campaña distinta, donde el eje X corresponde a la inversión en publicidad (USD) y el eje Y a las ventas generadas (USD).  

La **línea roja** corresponde a la línea de tendencia (regresión lineal), que permite visualizar de manera más clara el patrón general de los datos. En este caso, la pendiente positiva indica que **a mayor inversión en publicidad, mayor tiende a ser el ingreso por ventas**.  

Este resultado sugiere una relación **lineal y positiva** entre ambas variables, lo que significa que la inversión en marketing impacta directamente en el rendimiento económico de las campañas. No obstante, siempre es recomendable complementar este análisis con métricas estadísticas (como el coeficiente de correlación o R²) para medir la **fuerza** de dicha relación.


```python
# === Scatter + Trendline: Sales Revenue vs. Duration ===
x = marketing_df['Duration_Days']
y = marketing_df['Sales_Revenue_USD']

plt.figure(figsize=(6,4))
plt.scatter(x, y, label="Datos")

m, b = np.polyfit(x, y, 1)
plt.plot(x, m*x + b, color="red", label="Tendencia")

plt.xlabel("Duration (Days)")
plt.ylabel("Sales Revenue (USD)")
plt.title("Sales Revenue vs. Campaign Duration")
plt.legend()
plt.grid(True)
plt.show()
```

El gráfico de dispersión muestra la relación entre la **duración de la campaña (en días)** y los **ingresos en ventas (USD)**.  

Cada punto representa una campaña distinta. La **línea de tendencia** (en rojo) indica el comportamiento general de los datos. En este caso, la pendiente positiva refleja que, en promedio, **las campañas de mayor duración tienden a generar mayores ingresos en ventas**.  

Sin embargo, se observa que los puntos presentan más dispersión que en el caso de la relación con el gasto en publicidad. Esto sugiere que, aunque la duración tiene un efecto positivo sobre las ventas, **no es tan determinante como la inversión publicitaria**, ya que pueden influir otros factores como la plataforma utilizada, la estrategia de marketing o la calidad del anuncio.  

En conclusión, existe una relación positiva entre la duración de la campaña y los ingresos, pero no tan fuerte como la relación entre inversión y ventas.
